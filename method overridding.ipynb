```python
class bird():


    def __init__(self):
        self.value = "Inside bird"


    def show(self):
        print(self.value)

class sparrow(bird):


    def __init__(self):
        self.value = "Inside sparrow"


    def show(self):
        print(self.value)


obj1 = bird()
obj2 = sparrow()

obj1.show()
obj2.show()

```
